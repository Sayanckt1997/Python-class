### **Group By and Aggregate**
The groupby() function is used to group the DataFrame by one or more columns and then apply an aggregation function to each group.

```python
# Sample DataFrame with multiple departments
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
        'Department': ['HR', 'IT', 'HR', 'IT', 'HR'],
        'Age': [24, 19, 22, 32, 29],
        'Salary': [50000, 48000, 52000, 58000, 60000]}

df = pd.DataFrame(data)
df
```

### **Merging and Concating**

Merging and joining are essential techniques in Pandas for combining datasets.

```python
#pd.merge(left, right, how='inner', on=None)
#left and right: The two DataFrames to merge.
#how: Type of merge ('left', 'right', 'inner', 'outer').
#on: Column name(s) on which to merge. If not provided, the merge is based on common columns.
```

```python
# DataFrame 1
data1 = {'Employee': ['John', 'Alice', 'Bob', 'David'],
         'Department': ['HR', 'IT', 'Finance', 'IT'],
         'Age': [25, 30, 45, 35]}

df1 = pd.DataFrame(data1)

# DataFrame 2
data2 = {'Employee': ['John', 'Alice', 'Bob', 'Michael'],
         'Salary': [50000, 60000, 55000, 65000]}

df2 = pd.DataFrame(data2)

print(df1)
print(df2)
```

```python
# Merge DataFrames on 'Employee' column
df_merged = pd.merge(df1, df2, on='Employee', how='inner')
df_merged
```

```python
# Merge DataFrames on 'Employee' column
df_merged = pd.merge(df1, df2, on='Employee', how='outer')
df_merged
```

**Concatenating** in Pandas is used to combine two or more DataFrames either vertically (row-wise) or horizontally (column-wise). The **concat()** function in Pandas is used to perform concatenation.

```python
# DataFrame 1
data1 = {'Name': ['Alice', 'Bob'],
         'Department': ['HR', 'IT'],
         'Salary': [50000, 48000]}

df1 = pd.DataFrame(data1)

# DataFrame 2
data2 = {'Name': ['Charlie', 'David'],
         'Department': ['HR', 'IT'],
         'Salary': [52000, 58000]}

df2 = pd.DataFrame(data2)

print(df1)
print(df2)
```

```python
# Concatenate DataFrames row-wise
df_concat = pd.concat([df1, df2], axis=0, ignore_index=True)

df_concat
```

```python
# Concatenate DataFrames column-wise
# DataFrame 3
data3 = {'Age': [24, 29, 30, 35]}

df3 = pd.DataFrame(data3)

# Concatenate DataFrames column-wise
df_concat_col = pd.concat([df_concat, df3], axis=1)

df_concat_col
```

```python
# DataFrame with non-matching columns
data4 = {'Name': ['Eve'],
         'Department': ['HR'],
         'Experience': [5]}  # New column 'Experience'

df4 = pd.DataFrame(data4)

# Concatenate with non-matching columns
df_concat_mismatch = pd.concat([df_concat, df4], axis=0, ignore_index=True)

df_concat_mismatch
```

### **Applying functions**

The .apply() function in Pandas allows you to apply custom functions to DataFrames or Series, making it a powerful tool for data transformation. You can use it to apply a function to each element, row, or column.

```python
# Sample DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
        'Age': [24, 19, 22, 32, 29]}

df = pd.DataFrame(data)

df
```

```python
# Custom function to increase age by 1
def add_one(x):
    return x + 1

# Applying the function to the 'Age' column
df['Age'] = df['Age'].apply(add_one)

df
```

```python
#Using .apply() on a DataFrame (Row-wise or Column-wise):
# Function to create a new column with a message based on age
def age_group(row):
    if row['Age'] < 25:
        return 'Young'
    else:
        return 'Adult'

# Apply the function row-wise (axis=1)
df['Age Group'] = df.apply(age_group, axis=1)
df
```

Using .apply() with a Lambda Function

```python
# Use lambda function to square the 'Age' values
df['Age Squared'] = df['Age'].apply(lambda x: x ** 2)

df
```

```python
# Use lambda function for multi-condition logic
df['Age Category'] = df['Age'].apply(lambda x: 'Young' if x < 30 else ('Middle-Aged' if 30 <= x <= 40 else 'Old'))

df
```

```python
# Apply a string function to convert all names to uppercase
df['Name Upper'] = df['Name'].apply(str.upper)

df
```

```python
import random
from datetime import datetime, timedelta

# Function to generate sample data
def generate_sample_data(n_rows=100):
    random.seed(42)
    employees = ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank']
    departments = ['HR', 'IT', 'Sales', 'Marketing']
    start_date = datetime(2020, 1, 1)

    data = {
        #'Employee': [random.choice(employees) for _ in range(n_rows)],
        'Department': [random.choice(departments) for _ in range(n_rows)],
        'Sales': np.random.randint(20000, 100000, size=n_rows),
        'Profit': np.random.randint(2000, 20000, size=n_rows),
        'Quantity': np.random.randint(1, 100, size=n_rows),
        'Date': [start_date + timedelta(days=np.random.randint(0, 365)) for _ in range(n_rows)]
    }

    df = pd.DataFrame(data)
    return df

# Create sample DataFrame
df = generate_sample_data()
df
```

```python
# Line Plot using Matplotlib (Sales over Time)
# Sort by Date and plot Sales over Time
df_sorted = df.sort_values('Date')
plt.plot(df_sorted['Date'], df_sorted['Sales'], marker='o')
plt.title('Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Bar plot showing average profit by department
df.groupby('Department')['Profit'].mean().plot(kind='bar', color='green')
plt.title('Average Profit by Department')
plt.xlabel('Department')
plt.ylabel('Average Profit')
plt.grid(axis='y')
plt.show()
```

```python
# Scatter plot comparing Sales and Profit
plt.scatter(df['Sales'], df['Profit'], color='blue', alpha=0.5)
plt.title('Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.grid(True)
plt.show()
```

A **histogram** is a type of plot that represents the distribution of a dataset. It shows the frequency (or count) of data points that fall within a range of values, known as bins. Each bin represents an interval of values, and the height of the bar in the histogram indicates how many data points fall into that bin.

```python
# Histogram to visualize the distribution of Sales
df['Sales'].plot(kind='hist', bins=20, color='purple', edgecolor='black')
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Creating a scatter plot with a color bar for the 'Profit' column
scatter = plt.scatter(df['Sales'], df['Quantity'],
                      c=df['Profit'], cmap='viridis', s=100, alpha=0.6)

# Adding a color bar to the plot
plt.colorbar(scatter, label='Profit')

# Customizing the plot
plt.title('Sales vs Quantity with Color Bar for Profit')
plt.xlabel('Sales')
plt.ylabel('Quantity')
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
